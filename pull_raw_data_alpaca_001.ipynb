```python
from datetime import date, datetime, timedelta
from zoneinfo import ZoneInfo

from alpaca.data.historical.stock import StockHistoricalDataClient
from alpaca.data.requests import StockBarsRequest, StockTradesRequest, StockQuotesRequest
from alpaca.data.timeframe import TimeFrame, TimeFrameUnit
```

```python
# Define the stock historical data client
stock_historical_data_client = StockHistoricalDataClient('', '')  # Replace with API key and secret
```

```python
# Specify the symbols for market data
symbols = ['TSLA', 'GOOG', 'AMZN', 'MSFT', 'JPM', 'BA', 'UNH', 'XOM', 'GM']
#length of time in minutes for candlesticks
stick_width = 30
#number of days to look back on
look_back = 100

for symbol in symbols:
    # Define now based on the New York time zone
    now = datetime.now(ZoneInfo("America/New_York"))
    
    # Get historical bars by symbol
    # format API request
    # ref. https://docs.alpaca.markets/reference/stockbars-1
    req = StockBarsRequest(
        symbol_or_symbols = [symbol],
        timeframe=TimeFrame(amount = 30, unit = TimeFrameUnit.Minute), # specify timeframe
        start = now - timedelta(days = look_back),                          # specify start datetime, default=the beginning of the current day.
        #end=None,                                        # specify end datetime, default=now
        #limit = 2,                                               # specify limit
    )

    #send request to Alpaca, receive dataframe
    df = stock_historical_data_client.get_stock_bars(req).df
    #save dataframe to disk for future use
    file_name = symbol+'_'+str(look_back)+'_days_'+str(stick_width)+'_minute_candles.csv'
    df.to_csv(file_name)
```
